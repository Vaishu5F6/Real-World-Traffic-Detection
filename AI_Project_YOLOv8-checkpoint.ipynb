```python
from ultralytics import YOLO
from pathlib import Path
from PIL import Image, ImageEnhance
import numpy as np

import torch
```

```python
# Set up label encodings and data paths

label_encodings = ['bicycle', 'bus', 'car', 'motorbike', 'person']
data_paths = {
    "train": {
        "images": Path("./dataset/train/images"),
        "labels": Path("./dataset/train/labels"),
    },
    "test": {
        "images": Path("./dataset/test/images"),
        "labels": Path("./dataset/test/labels"),
    },
    "val": {
        "images": Path("./dataset/val/images"),
        "labels": Path("./dataset/val/labels"),
    },
}
```

```python
model = YOLO('yolov8n.yaml').load('yolov8n.pt')
yaml_path = Path('./dataset/data.yaml').resolve()
results = model.train(data=str(yaml_path), epochs=60, imgsz=640)
```

```python
# results = model.train(data=str(yaml_path), epochs=60, batch=32, imgsz=640, resume=True)
```

```python
metrics = model.val()
```

```python
conf_mat = metrics.confusion_matrix
total_events = sum(sum(conf_mat.matrix))

tp, fp = conf_mat.tp_fp()
accuracy = sum(tp) / total_events
accuracy
```
